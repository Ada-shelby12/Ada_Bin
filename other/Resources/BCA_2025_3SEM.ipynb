```python
"""
Q1: Write a Python program to simulate a Simple Reflex Agent (decisions based solely on current percept)
for a two-room vacuum cleaner world.  The agent should sense the state of the room (clean or dirty),
act accordingly, and continue until all rooms are clean.

"""

def perceive(environment, location):
    # Get the status of the current room
    return environment[location]

def move(location):
    # Switch location between A and B
    return "B" if location == "A" else "A"

def act(environment, location):
    status = perceive(environment, location)
    if status == "Dirty":
        print(f"Room {location} is Dirty. Action: Clean.")
        environment[location] = "Clean"
    else:
        print(f"Room {location} is already Clean. Action: Move to next room.")
        location = move(location)
    return environment, location

def run():
    environment = {"A": "Dirty", "B": "Dirty"}
    location = "A"
    print("Initial Environment State:", environment)
    print("\nAgent started cleaning...\n")
    steps = 0

    while "Dirty" in environment.values():
        print(f"Step {steps + 1}: Agent in Room {location}")
        environment, location = act(environment, location)
        steps += 1

    print("\nFinal Environment State:", environment)
    print(f"Cleaning completed in {steps} steps.")

if __name__ == "__main__":
    run()



```

```python
"""
Q4: Implement the Breadth-First Search (BFS) algorithm to find the shortest path between two nodes in an unweighted graph
and print the shortest path from the start node to the goal node.
"""

# Import deque from collections to use a queue for BFS.
from collections import deque

# Define the bfs function to find the shortest path between start and goal nodes in a graph.
def bfs(graph, start, goal):
    """
    Perform BFS to find the shortest path between the start and goal nodes in a graph.

    Parameters:
    - graph: dict, adjacency list representation of the graph.
    - start: int, the starting node.
    - goal: int, the target node.

    Returns:
    - list: the shortest path from start to goal, or None if no path exists.
    """
    # Handle the edge case where the start and goal nodes are the same.
    if start == goal:
        return [start]

    # Initialize a queue with the starting node as the initial path.
    queue = deque([[start]])

    # Initialize a set to keep track of visited nodes to prevent revisiting them.
    visited = set()

    # Perform BFS until the queue is empty.
    while queue:
        # Dequeue the first path from the queue.
        path = queue.popleft()

        # Get the last node from the current path (this is the node being explored).
        node = path[-1]

        # If the node has not been visited yet, explore its neighbors.
        if node not in visited:
            # Mark the current node as visited.
            visited.add(node)

            # Iterate over all neighbors of the current node.
            for neighbor in graph.get(node, []):  # Use graph.get() to handle missing nodes gracefully.
                # Create a new path by extending the current path with the neighbor.
                new_path = list(path)
                new_path.append(neighbor)

                # Add the new path to the queue for further exploration.
                queue.append(new_path)

                # If the neighbor is the goal, return the path leading to it.
                if neighbor == goal:
                    return new_path

    # If no path is found after exploring all possibilities, return None.
    return None


# Example graph: Adjacency list representation.
graph = {
    1: [2, 3],      # Node 1 is connected to nodes 2 and 3
    2: [1, 4, ],   # Node 2 is connected to nodes 1, 4, and 5
    3: [1, 6],      # Node 3 is connected to nodes 1 and 6
    4: [2, 5],         # Node 4 is connected to node 2
             # Node 5 is connected to node 2
    6: [3, 7],      # Node 6 is connected to nodes 3 and 7
    7: [6]          # Node 7 is connected to node 6
}

# Find the shortest path from node 1 to node 5 and print the result.
print("Shortest path from node 1 to node 5:", bfs(graph, 1, 5))

# Find the shortest path from node 1 to node 7 and print the result.
print("Shortest path from node 1 to node 7:", bfs(graph, 1, 7))

```

```python
"""
Q5: Write a Python program to apply the AO* algorithm on a given AND-OR graph to determine the minimal
cost solution path from the start node to terminal nodes.
"""


graph = {
    'A': [['B', 'C'], ['D']],  # OR node with 2 options: AND(B,C) or OR(D)
    'B': [['E']],              # AND node with 1 group: E
    'C': [['F']],              # AND node with 1 group: F
    'D': [],                   # Terminal
    'E': [],                   # Terminal
    'F': []                    # Terminal
}

heuristic = {
    'A': 10,
    'B': 4,
    'C': 2,
    'D': 3,
    'E': 0,
    'F': 0
}

solution_graph = {}
solved = {}

def ao_star(node):
    # If node is terminal
    if not graph[node]:
        solved[node] = True
        return heuristic[node]

    # If already solved
    if solved.get(node, False):
        return heuristic[node]

    min_cost = float('inf')
    best_option = None

    # Explore each group (AND group)
    for group in graph[node]:
        cost = 0
        # Sum costs of children in the AND group
        for child in group:
            cost += ao_star(child)
        if cost < min_cost:
            min_cost = cost
            best_option = group

    # Update heuristic and solution graph if cost improved
    if heuristic[node] != min_cost:
        heuristic[node] = min_cost
        solution_graph[node] = best_option

    # Check if all children in best option are solved
    if all(solved.get(c, False) for c in best_option):
        solved[node] = True

    return heuristic[node]

# Run AO* from root node
ao_star('A')

# Print solution graph
print("\nSolution graph:")
for parent, children in solution_graph.items():
    print(f"{parent} -> {children}")

print("\nHeuristics:")
for node, cost in heuristic.items():
    print(f"{node}: {cost}")

print("\nSolved nodes:")
print(solved)
```

```python
"""
Q6: Apply the A* Search algorithm to find the shortest path in a 4x4 grid.
"""

"""
Problem Setup
Grid size: 4x4

Start: Top-left corner (0,0)

Goal: Bottom-right corner (3,3)

Obstacles: Can be added optionally (for now, assuming all cells are walkable)

Movement: Up, down, left, right (no diagonals)
"""



import heapq

# Define the grid size
ROWS, COLS = 4, 4

# Directions: up, down, left, right
DIRS = [(-1, 0), (1, 0), (0, -1), (0, 1)]

def heuristic(a, b):
    # Manhattan distance
    return abs(a[0] - b[0]) + abs(a[1] - b[1])

def a_star_search(start, goal):
    open_set = []
    heapq.heappush(open_set, (0 + heuristic(start, goal), 0, start))

    came_from = {}
    g_score = {start: 0}

    while open_set:
        _, cost, current = heapq.heappop(open_set)

        if current == goal:
            # Reconstruct path
            path = []
            while current in came_from:
                path.append(current)
                current = came_from[current]
            path.append(start)
            path.reverse()
            return path

        for dx, dy in DIRS:
            neighbor = (current[0] + dx, current[1] + dy)

            if 0 <= neighbor[0] < ROWS and 0 <= neighbor[1] < COLS:
                tentative_g = g_score[current] + 1
                if neighbor not in g_score or tentative_g < g_score[neighbor]:
                    g_score[neighbor] = tentative_g
                    priority = tentative_g + heuristic(neighbor, goal)
                    heapq.heappush(open_set, (priority, tentative_g, neighbor))
                    came_from[neighbor] = current

    return None  # No path found

# Example usage
start = (0, 0)
goal = (3, 3)
path = a_star_search(start, goal)

if path:
    print("Shortest path from", start, "to", goal, ":")
    for step in path:
        print(step)
else:
    print("No path found.")
```

```python
"""
Q7: Implement the Simple Hill Climbing algorithm to find a local maximum in a 1D list of numeric values, where a neighbor is defined
as an adjacent element. Display the steps taken to reach the peak.
"""


def simple_hill_climb_1d(data, start_index=0):
    """
    Simple Hill Climbing in a 1D list.
    At each step, move to the higher neighbor if available.
    Stops at a local maximum.
    """

    # Start from the given index
    current = start_index

    # Keep track of the path (indices visited)
    path = [current]

    while True:
        # Determine the indices of the neighbors
        left = current - 1 if current > 0 else None    # Left neighbor (if within bounds)
        right = current + 1 if current < len(data) - 1 else None  # Right neighbor (if within bounds)

        # List to store available neighbors and their values
        neighbors = []

        # Add valid left neighbor to the list
        if left is not None:
            neighbors.append((left, data[left]))

        # Add valid right neighbor to the list
        if right is not None:
            neighbors.append((right, data[right]))

        # If there are no neighbors (shouldn’t happen in normal cases), stop
        if not neighbors:
            break

        # Find the neighbor with the highest value
        best_neighbor = max(neighbors, key=lambda x: x[1])  # x[0]=index, x[1]=value

        # Move to neighbor only if it's better (higher value)
        if best_neighbor[1] > data[current]:
            current = best_neighbor[0]      # Update current index to better neighbor
            path.append(current)            # Record the move
        else:
            break  # No better neighbor found — local maximum reached

    # Final output
    print("Data List: ", data)
    print("Visited Indices: ", path)
    print("Visited Values: ", [data[i] for i in path])
    print("Local Maximum Value:", data[current], "at index", current)


# Example usage
data = [1, 2, 3, 5, 6, 2, 1]
simple_hill_climb_1d(data, start_index=0)


```

```python
"""
Implement a solution to the 4-Queens problem using backtracking as a Constraint Satisfaction Problem (CSP), where variables represent queen positions on the board, and constraints ensure that no two queens attack each other (row, column, and diagonal constraints).
"""

def is_safe(queen_positions, row, col):
    """
    Check if placing a queen at (row, col) is safe from attacks.
    It is unsafe if:
      - Any previously placed queen is in the same column
      - Any previously placed queen is on the same diagonal
    """
    for r in range(row):                         # Check all previously placed queens (rows 0 to row-1)
        c = queen_positions[r]                   # Column of queen in row r
        if c == col:                             # If same column as current position
            return False  #  same column
        if abs(r - row) == abs(c - col):        # If on the same diagonal (difference in rows == difference in columns)
            return False  #  same diagonal
    return True  #  safe to place  # No conflicts found, safe to place queen here


def solve_queens(n, row=0, queen_positions=None):
    """
    Solve the N-Queens problem using backtracking (CSP-style).
    Arguments:
      n: board size (N)
      row: current row being filled
      queen_positions: current partial assignment (index = row, value = column)
    Returns:
      List of valid solutions (each solution is a list of queen positions)
    """
    if queen_positions is None:                  # If no partial solution provided
        queen_positions = [-1] * n               # Initialize the board with -1 meaning no queen placed yet in any row

    if row == n:                                 # If all rows are filled with safe queen positions
        return [queen_positions.copy()]          # Return a copy of this complete solution as a valid one

    solutions = []                               # Initialize list to collect all solutions from this branch
    for col in range(n):                         # Try placing queen in every column in the current row
        if is_safe(queen_positions, row, col):  # Check if position (row, col) is safe
            queen_positions[row] = col          # Place queen at (row, col)
            solutions += solve_queens(n, row + 1, queen_positions)  # Recurse to next row to place next queen
            queen_positions[row] = -1           # Backtrack: Remove queen from (row, col) to try next position

    return solutions                             # Return all collected solutions


# Solve the 4-Queens problem
solutions = solve_queens(4)                     # Call solver for a 4x4 board

# Print each solution as a board
for sol in solutions:                            # For each solution (list of queen columns by row)
    print("Solution:")                           # Print header
    for row in range(4):                         # For each row on the board
        line = ""                                # Start with empty string for this row
        for col in range(4):                      # For each column in this row
            if sol[row] == col:                   # If queen placed in this column
                line += "Q "                      # Add 'Q' for queen
            else:
                line += ". "                      # Add '.' for empty space
        print(line)                               # Print the constructed row
    print()                                      # Print blank line after each solution
```

```python
"""
  Q9: Design and implement an AI-based disease diagnosis expert system using Python. The system must:
• Use rule-based logic (dictionary-based or with a rule engine).
• Support 5 diseases.
• Include 10 symptom-based rules mapping symptoms to diseases.
• Accept user-input symptoms.
• Infer all possible matching diseases based on a symptom match score (e.g., percentage match).
• Allow dynamic addition of a new disease and associated symptoms at runtime.
• Rerun inference with the updated rules.

"""

# Rule-based Disease Diagnosis Expert System (Scoring-Based Matching)

def diagnose(symptoms, rules, threshold=0.5):
    matched_diseases = []
    for disease, rule_symptoms in rules.items():
        matched_count = len(set(symptoms).intersection(rule_symptoms))
        total_required = len(rule_symptoms)
        match_score = matched_count / total_required if total_required > 0 else 0

        if match_score >= threshold:
            matched_diseases.append((disease, round(match_score, 2)))
    return matched_diseases

def add_disease(rules):
    name = input("Enter new disease name: ").strip()
    symptoms = input(f"Enter symptoms for {name} (comma-separated): ").lower().split(',')
    symptoms = [s.strip() for s in symptoms if s.strip()]
    rules[name] = symptoms
    print(f"Disease '{name}' added successfully with symptoms: {symptoms}")

# Initial rule base: 5 diseases, 10 unique symptoms across rules
disease_rules = {
    'Flu': ['fever', 'cough'],
    'Cold': ['sneezing', 'runny nose'],
    'Malaria': ['fever', 'chills'],
    'Migraine': ['headache', 'nausea'],
    'Allergy': ['rash', 'itching']
}

# Accept patient symptoms
input_symptoms = input("Enter your symptoms (comma-separated): ").lower().split(',')
patient_symptoms = [s.strip() for s in input_symptoms if s.strip()]

# Run inference
print("\nPossible diagnoses based on your symptoms:")
matches = diagnose(patient_symptoms, disease_rules, threshold=0.5)
if matches:
    for disease, score in matches:
        print(f" - {disease} (match: {int(score * 100)}%)")
else:
    print("No significant matches found.")

# Allow dynamic addition of a new disease
add_new = input("\nDo you want to add a new disease? (yes/no): ").strip().lower()
if add_new == 'yes':
    add_disease(disease_rules)
    # Rerun inference
    print("\nUpdated diagnosis after adding new disease:")
    matches = diagnose(patient_symptoms, disease_rules, threshold=0.5)
    if matches:
        for disease, score in matches:
            print(f" - {disease} (match: {int(score * 100)}%)")
    else:
        print("No significant matches found.")
```

```python
"""
Q10. Using Python NLTK, perform the following Natural Language Processing (NLP) tasks for text content.
a) Tokenizing
b) Filtering Stop Words
c) Stemming
d) Part of Speech tagging
e) Chunking
f) Named Entity Recognition (NER)
"""
# Adjust the code for any nltk issues

import nltk
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger_eng')
nltk.download('maxent_ne_chunker_tab')
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk import pos_tag, ne_chunk

# Download required resources (run once)
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
nltk.download('words')

text = "Apple Inc. is looking at buying U.K. startup for $1 billion. This is a great opportunity."

# a) Tokenizing
print("a) Tokenizing")
# Sentence tokenization splits text into sentences
sentences = sent_tokenize(text)
print("Sentences:", sentences)
# Word tokenization splits text into individual words/tokens
words = word_tokenize(text)
print("Words:", words)
print()

# b) Filtering Stop Words
print("b) Filtering Stop Words")
stop_words = set(stopwords.words('english'))
# Remove common English stopwords from tokenized words
filtered_words = [w for w in words if w.lower() not in stop_words]
print("Filtered Words (no stopwords):", filtered_words)
print()

# c) Stemming
print("c) Stemming")
ps = PorterStemmer()
# Reduce words to their word stem/root form
stemmed_words = [ps.stem(w) for w in filtered_words]
print("Stemmed Words:", stemmed_words)
print()

# d) Part of Speech Tagging
print("d) Part of Speech Tagging")
# Assign grammatical part of speech tags to each word token
pos_tags = pos_tag(words)
print("POS Tags:", pos_tags)
print()

# e) Chunking
print("e) Chunking")
# Define simple noun phrase (NP) chunk grammar
grammar = "NP: {<DT>?<JJ>*<NN.*>+}"
cp = nltk.RegexpParser(grammar)
# Parse POS-tagged words to identify noun phrase chunks
chunk_tree = cp.parse(pos_tags)
print("Chunk Tree:")
print(chunk_tree)
print()

# f) Named Entity Recognition (NER)
print("f) Named Entity Recognition")
# Identify named entities (persons, organizations, locations, etc.)
ner_tree = ne_chunk(pos_tags)
print("Named Entities Tree:")
print(ner_tree)
print()
```
