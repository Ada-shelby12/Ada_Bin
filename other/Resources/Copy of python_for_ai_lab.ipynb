```python
# Q3: How do you use a priority queue to always explore the lowest-cost node first?

import heapq

open_set = []
start = (0, 0)
goal = (3, 3)
initial_cost = 0
heapq.heappush(open_set, (initial_cost + heuristic(start, goal), initial_cost, start))

# Extract the node with the lowest priority
print(heapq.heappop(open_set))
```

```python
# Q2: How do you detect row or diagonal conflicts between two queens?

def has_conflict(col1, row1, col2, row2):
    same_row = row1 == row2
    same_diag = abs(row1 - row2) == abs(col1 - col2)
    return same_row or same_diag

# Check conflict between Q1 at (0, 1) and Q2 at (2, 3)
print(has_conflict(0, 1, 2, 3))
```

```python
# Q3: How do you use backtracking to explore different placements of queens?

def try_placing_queen(col):
    if col == 4:
        print("Solution found!")
        return True  # All columns filled

    for row in range(4):
        print(f"Trying queen at column {col}, row {row}")
        # Assume it's valid and recurse (in real code, you'd check conflicts here)
        if try_placing_queen(col + 1):
            return True

    return False  # Trigger backtracking

try_placing_queen(0)
```

```python
# Q7: How do you reuse the diagnosis function after updating rules?

def diagnose(symptoms, rules):
    matches = []
    for disease, known in rules.items():
        score = len(set(symptoms) & set(known)) / len(known)
        if score >= 0.5:
            matches.append((disease, score))
    return matches

# Reuse diagnose with new rule
print(diagnose(['fever', 'fatigue'], rules))
```

```python
# Why use a heap (via heapq) for the open set?

# Purpose: To efficiently get the node with the lowest cost (f = g + h) during A* search.

# Why a heap?
# A regular list would require O(n) time to find the min element. A heap gives O(log n) push/pop for best-node selection.


import heapq

# Min-heap to always get the node with lowest f-score
open_set = []

# (f_score, g_score, node)
heapq.heappush(open_set, (5, 1, (0, 0)))
heapq.heappush(open_set, (3, 2, (1, 0)))
heapq.heappush(open_set, (4, 2, (0, 1)))

# Always pops the lowest f_score node
print(heapq.heappop(open_set))  # Output: (3, 2, (1, 0))

```

```python
#  Why use a dictionary for g_score?

# Purpose: To track the shortest path cost from the start to each node.

# Why a dictionary?
# You need O(1) lookup to compare and update path costs per node. A list would be too slow or memory-inefficient for arbitrary grid positions.

g_score = {}

# First time visiting a node
g_score[(0, 0)] = 0

# Update with a better path
g_score[(1, 0)] = 1

# Check if a shorter path is found
new_g = 2
if (1, 0) not in g_score or new_g < g_score[(1, 0)]:
    g_score[(1, 0)] = new_g

print("g_score dictionary:", g_score)

```

```python
# Question: What does this code do, and why is `came_from` a dictionary?
# Answer:
# The code reconstructs the shortest path from a goal node `(2, 0)` to the start node `(0, 0)`
# by following each node's parent stored in the `came_from` dictionary.
# It builds the path in reverse, then reverses it to get the correct order.
# `came_from` is a dictionary that maps each node to the node it came from.

# ==== CODE ====

came_from = {}

# Record how you reached each node
came_from[(1, 0)] = (0, 0)
came_from[(2, 0)] = (1, 0)

# Reconstruct path from goal to start
current = (2, 0)
path = []
while current in came_from:
    path.append(current)
    current = came_from[current]

path.append((0, 0))  # Start node
path.reverse()

print(path)  # Output: [(0, 0), (1, 0), (2, 0)]
```
